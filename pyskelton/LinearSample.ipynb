```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pyDOE2 as doe
```

```python
lhs = doe.lhs(5, samples=100, criterion='center',random_state=0)
min_values = np.array([100, 1000, 1, 20, 50])
max_values = np.array([0, 500, 0, 10, 40])

df = pd.DataFrame(
    min_values + lhs*(max_values - min_values), 
    columns=[
        "Recipe A(setting data)", 
        "Recipe B(setting data)", 
        "Recipe C(setting data)",
        "Recipe D(setting data)", 
        "Recipe E(setting data)", 
    ]
)
df
```

```python
# A
er = np.random.randn(5,100)
df["Sensor A"] = df["Recipe A(setting data)"] + (max_values[0] + min_values[0])/2 * er[0]*0.01
df["Sensor B"] = df["Recipe B(setting data)"] + (max_values[1] + min_values[1])/2 * er[1]*0.01
df["Sensor C"] = df["Recipe C(setting data)"] + (max_values[2] + min_values[2])/2 * er[2]*0.01
df["Sensor D"] = df["Recipe D(setting data)"] + (max_values[3] + min_values[3])/2 * er[3]*0.01
df["Sensor E"] = df["Recipe E(setting data)"] + (max_values[4] + min_values[4])/2 * er[4]*0.01

df["Sensor F"] = 0.3 * df["Sensor A"] - 0.01 * df["Sensor B"] + 10.0 * df["Sensor C"]
df["Sensor G"] = 0.05 * df["Sensor B"] - 3.0 * df["Sensor D"]
df["Sensor H"] = 0.5 * df["Sensor D"] + 0.01 * df["Sensor E"]
df["Sensor I"] = 12.0 * df["Sensor F"] - 0.02 * df["Sensor E"]

df["Measure A(quality data)"] = 2.0 * df["Sensor H"] - 0.1 * df["Sensor A"] 

df
```

```python
df.to_csv('data/linear_sample.csv', index=False)
```
